```python
kafkaHosts = KAKFA_HOST_MAP.get('prod_external', [])
KafkaTopic = KAFKA_CRAWELD_PAGE['KafkaTopic']

print(kafkaHosts)
print(KafkaTopic)
```

查看已经爬取的page 页

```python
setting = {
    'KafkaHost': kafkaHosts,
    'KafkaTopic': KafkaTopic,
    'NextMainJobConsumeGroup': 'next-main-test',
    'AdditionalFieldJobGroup': 'additional-test',
    'MaxPollRecords': 10000
}

connector = KafkaSpiderStatusConnector(kafkaHosts, {'KAFKA_CRAWELD_PAGE': setting})
connector.full_consumer.consumer.partitions_for_topic(KafkaTopic)
```

```python
page = connector.get_crawled_page()
print(len(page))
```
